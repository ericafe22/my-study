# 決定木とは  
意味が解釈しやすいモデル  
特徴量に基づく質問を学習しクラスを推測する  
情報利得(information gain)が最大となる特徴量でデータを分割する  

```python
# 必要そうなパッケージをimport
from sklearn import datasets
import numpy as np
import pandas as pd
from scipy import interp
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import learning_curve, validation_curve
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc
from sklearn.pipeline import make_pipeline
from pydotplus import graph_from_dot_data
from IPython.display import Image
```

```python
# 結果を図示するための関数を定義
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        # plot all samples
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')
```

```python
# サンプルのIrisデータセットをロード
iris = datasets.load_iris()
# 3, 4列目の特徴量を抽出
X = iris.data[:, [2, 3]]
# クラスラベルを取得
y = iris.target
# トレーニングデータとテストデータに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1, stratify=y)

# 説明変数を標準化する
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
#criterion：分割する際に利用する品質。ジニ係数を利用する “gini” と情報量を示す “entropy” のいずれかを指定
#max_depth：枝の深さ
tree = DecisionTreeClassifier(criterion='gini', 
                              max_depth=4, 
                              random_state=1)
tree.fit(X_train_std, y_train)
```

```python
# treeの図示
dot_data = export_graphviz(tree,
                           filled=True, 
                           rounded=True,
                           class_names=['Setosa', 
                                        'Versicolor',
                                        'Virginica'],
                           feature_names=['petal length', 
                                          'petal width'],
                           out_file=None) 
graph = graph_from_dot_data(dot_data) 
graph.write_png('tree.png')
Image(graph.create_png())
```

```python
# 予想結果の図示
X_combined = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X_combined, y_combined, 
                      classifier=tree, test_idx=range(105, 150))

plt.xlabel('petal length [cm]')
plt.ylabel('petal width [cm]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 予測結果
tree.predict(X_test_std)
```

```python
# 一致状況
print(y_test == tree.predict(X_test_std))
```

```python
# 正解率
accuracy_score(y_test, tree.predict(X_test_std))
```

# 分類モデルに対する色々な評価値

### 事前準備

```python
# 2値分類のデータセットを用意
# 癌の予測用のデータセット
# https://www.kaggle.com/uciml/breast-cancer-wisconsin-data

df = pd.read_csv('https://archive.ics.uci.edu/ml/'
                 'machine-learning-databases'
                 '/breast-cancer-wisconsin/wdbc.data', header=None)
df.head(5)
```

```python
print(df.shape)
df.head()
```

```python
# 特徴量とtargetの取り出し
X = df.loc[:, 2:].values
y = df.loc[:, 1].values
# B, Mをそれぞれ0, 1に置き換える
le = LabelEncoder()
y = le.fit_transform(y)
le.classes_
```

```python
# training setとtest setの分割
X_train, X_test, y_train, y_test =     train_test_split(X, y, 
                     test_size=0.20,
                     stratify=y,
                     random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
# ロジスティック回帰による分類
# パイプラインにより標準化とロジスティック回帰を一度に行う
# make-pipeline：標準化+ロジスティック回帰
pipe_lr = make_pipeline(StandardScaler(),
                        LogisticRegression(penalty='l2', random_state=1 ,solver='lbfgs'))
pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
y_pred
```

### 正解率  
テストデータセットがどれだけ正しく分類できたかを表す。  
例えば100個のテストデータセットで1個だけの分類誤りであれば、正解率は0.99  
基本的にはこの指標だけで評価できるが一部例外もある

```python
print('Accuracy: {}'.format(accuracy_score(y_test, y_pred)))
```

### 混同行列

(再掲)



|  名前  |  算出式 |  概要 |
| :-- | :-- | :-- |
|正解率 (accuracy)|(TP + TN) / (TP + FP + TN + FN)|  分類の正解率。全体のうちでどれだけが正しく分類できたか。|
|誤分類率(Error)|(FP + FN) / (TP + FP + TN + FN)|  分類の誤分類率率。全体のうちでどれだけが誤って分類したか。|
|真陽性率(TPR)/再現率/検出率 (Recall)|TP / (TP + FN)|  陽性のデータのうち、どれだけを陽性と分類できたか|
|-| FN / (TP + FN)| 陽性のデータのうち、どれだけを陰性と判断したか|
|特異性 (specificity)|TN / (FP + TN)|  陰性のデータのうち、どれだけを陰性と分類できたか|
|偽陽性率(FPR)|FP / (FP + TN)|  陰性のデータのうち、どれだけを陽性と判断したか|
|適合率 (precision)|TP / (TP + FP)|  陽性と判断されたデータのうち、どれだけが正しく分類できたか|
| - |TN / (TN + FN)|陰性と判断されたデータのうち、どれだけが正しく分類できたか|


例えば、癌に罹った患者はそうでない患者に比べて圧倒的に少ない。  
このような場合に正解率でそのモデルの評価をすると、  
実際に癌の患者を正しく診断できなくとも高い正解率が出ることがある。  
しかし、実際には正しく癌の患者を診断できたかが重要である場合があるので、   
$TPR$での評価を行う。  

```python
# 混同行列の図示
confmat = confusion_matrix(y_true=y_test, y_pred=y_pred)
print(confmat)

fig, ax = plt.subplots(figsize=(2.5, 2.5))
ax.matshow(confmat, cmap=plt.cm.Blues, alpha=0.3)
for i in range(confmat.shape[0]):
    for j in range(confmat.shape[1]):
        ax.text(x=j, y=i, s=confmat[i, j], va='center', ha='center')

plt.xlabel('Predicted label')
plt.ylabel('True label')

plt.tight_layout()
plt.show()
```

### 再現率(Recall)と 適合率(Precision)と、F1スコア
正解率だけで判断しきれない場合には適合率・再現率・F1値の評価値でモデルを判断する


##### 再現率・検出率(Recall)
真が陽性のデータのうち、どれだけを陽性と分類できたか
$$Recall = \frac{TP}{TP+FN}$$

##### 適合率(Precision)
陽性と判断されたデータのうち、どれだけが正しく分類できたか
$$Precision = \frac{TP}{TP+FP}$$

##### F1値
F-measure, F-score, F1 Score とも呼ばれる。適合率と再現率の調和平均。  
検出率は高くても精度が低くなっていないか、逆に、精度は高くても検出率が低いモデルでないか、といった評価を示す。

$$F1 = 2\frac{PRE*REC}{PRE+REC}$$

```python
print('Precision: %.3f' % precision_score(y_true=y_test, y_pred=y_pred))
print('Recall: %.3f' % recall_score(y_true=y_test, y_pred=y_pred))
print('F1: %.3f' % f1_score(y_true=y_test, y_pred=y_pred))
```

### ROC曲線

#### 受信者操作特性(Reciever Operating Characteristic: ROC)曲線  
ROC曲線の曲面下面積(Area Under the Curve: AUC)を評価指標として計算できる。  
分類器のしきい値を変えることによって変動する 以下の2軸でグラフを描く 

真陽性率(TPR)：TP / (TP + FN)：真が陽性のデータのうち、どれだけを陽性と分類できたか  
偽陽性率(FPR)：FP / (FP + TN)：真が陰性のデータのうち、どれだけを陽性と判断したか  

```python
pipe_lr = make_pipeline(StandardScaler(),
                        LogisticRegression(penalty='l2', random_state=1,solver='lbfgs'))

# 説明変数の4列目と14列目を抜き出す
X_train2 = X_train[:, [4, 14]]
    
# 目的変数のラベル比率を保ったままデータをK個に分割
# testデータをブロックごとにスライドさせて全てのブロックがテストデータを経験するようなデータセットをK個分作成する
# https://medium.com/music-and-technology/scikit-learn%E3%81%AEstratifiedkfold%E3%81%A7%E4%BA%A4%E5%B7%AE%E6%A4%9C%E5%AE%9A-fcb3c9f4572b
cv = list(StratifiedKFold(n_splits=3, 
                          random_state=1).split(X_train, y_train))

fig = plt.figure(figsize=(7, 5))

mean_tpr = 0.0
mean_fpr = np.linspace(0, 1, 100)
all_tpr = []

for i, (train, test) in enumerate(cv):
    # ロジスティック回帰を実施
    probas = pipe_lr.fit(X_train2[train],
                         y_train[train]).predict_proba(X_train2[test])

    # TPR・FPR値を計算
    fpr, tpr, thresholds = roc_curve(y_train[test],
                                     probas[:, 1],
                                     pos_label=1)
    mean_tpr += interp(mean_fpr, fpr, tpr)
    mean_tpr[0] = 0.0

    # AUC値を計算
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr,
             tpr,
             label='ROC fold %d (area = %0.2f)'
                   % (i+1, roc_auc))

plt.plot([0, 1],
         [0, 1],
         linestyle='--',
         color=(0.6, 0.6, 0.6),
         label='random guessing')

mean_tpr /= len(cv)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
plt.plot(mean_fpr, mean_tpr, 'k--',
         label='mean ROC (area = %0.2f)' % mean_auc, lw=2)
plt.plot([0, 0, 1],
         [0, 1, 1],
         linestyle=':',
         color='black',
         label='perfect performance')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.legend(loc="lower right")

plt.tight_layout()
plt.show()
```

## 過学習に対する対処

**学習不足**(underfitting)とはトレーニングデータセットのパターンをうまく補足するにはモデルの複雑さが十分ではなく、未知のデータに対する性能が低いことを意味する。  
このようなモデルは「**バイアスが高い**」(high bias)とも表現される。

---

**過学習**(overfitting)は、トレーニングデータセットではうまく機能するモデルが、未知のデータ(テストデータセット)ではうまく汎化されないという問題である。  
過学習が発生しているモデルは「**バリアンスが高い**」(high variance)とも表現される。  
その原因として、パラメータの数が多すぎるために、データに対してモデルが複雑すぎることが考えられる。

---

このバリアンスとバイアスはトレードオフの関係にあることが知られている。


### 学習曲線について

トレーニングデータセットのサイズの関数として、モデルの学習データと検証データの正解率をプロットすると   
・モデルのバリアンスやバイアスが高いかどうか  
・データをさらに集めることが問題の解決に役立つかどうか  
を見ることができる。これを**学習曲線**という。

```python
# 学習曲線のプロット
pipe_lr = make_pipeline(StandardScaler(),
                        LogisticRegression(penalty='l2', random_state=1,solver='lbfgs'))

# Trainデータセットを分割して、学習・テストの比率を変えて、学習データとテストデータの正答率の変化をみていく
# トレーニングサンプル数を0.1倍から1倍（そのまま）まで10段階で変えつつ性能を評価
# 訓練データの正答率：training accurancy
# 検証データの正答率：validation accurancy

train_sizes, train_scores, test_scores = learning_curve(estimator=pipe_lr,
                               X=X_train,
                               y=y_train,
                               train_sizes=np.linspace(0.1, 1.0, 10),
                               cv=10,
                               n_jobs=1)

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.plot(train_sizes, train_mean,
         color='blue', marker='o',
         markersize=5, label='training accuracy')

plt.fill_between(train_sizes,
                 train_mean + train_std,
                 train_mean - train_std,
                 alpha=0.15, color='blue')

plt.plot(train_sizes, test_mean,
         color='green', linestyle='--',
         marker='s', markersize=5,
         label='validation accuracy')

plt.fill_between(train_sizes,
                 test_mean + test_std,
                 test_mean - test_std,
                 alpha=0.15, color='green')

plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim([0.8, 1.03])
plt.tight_layout()
plt.show()
```

上記の例だと  
トレーニング時のサンプルの個数が250個を超えている場合、このモデルの性能はトレーニングでもテストでも良い。  
またサンプルの個数が250個未満だと、過学習の度合いが高まっていることがわかる。

### 学習曲線によって偏っていることが判明した場合の対処法

#### バイアスが高い場合

* モデルのパラメータの個数を増やす
* 正則化のパラメータを減らすなどして、モデルの複雑さをあげる

#### バリアンスが高い場合

* トレーニングデータをさらに収集する
* 正則化のパラメータを増やすなどして、モデルの複雑さを抑える
* 特徴を適切に選ぶことで特徴量の個数を減らし、モデルの複雑さを抑える


トレーニング時のサンプルの個数が**250個を超えている場合**、このモデルの性能はトレーニングでもテストでも良い。  
また、サンプルの個数が**250個未満**だと、過学習の度合いが高まっていることがわかる。

### 正則化による過学習への対処

バイアスとバリアンスのトレードオフに対処する方法の一つとして、**正則化**に基づいてモデルの複雑さを調整することができる。  
L1、L2正則化を用いた機械学習モデルは損失関数とパラメータの値の和を最小にするパラメータを学習する。  
パラメータの和を評価することでパラメータが大きくなるのを防ぐ。

**正則化なし**  
損失関数

$$min f(x)$$

**L1正則化**  
損失関数+L1正則化項(パラメータの絶対値の和)  
余分な説明変数を省くことを目的として手法。次元圧縮のために用いられることが多い

$$min f(x) + {\lambda}\sum_{i=1}^m|w_i| = min f(x) + {\lambda}|w|$$


**L2正則化**  

損失関数+L2正則化項(パラメータの2乗和)  
過学習を防ぐことで精度を高めるために用いられる。結果としてL1正則化よりも予測精度が高い傾向にある

$$minf(x) + \frac{\lambda}{2}\sum_{j=1}^n|w_j|^2 = min f(x) + \frac{\lambda}{2}||w||^2$$

このとき、$\lambda$は正則化パラメータとなっており、これを変動させることでモデルの複雑さを調整することができる。  
正則化の背景にある考え方は、「*極端なパラメータの重みにペナルティを課す*」ということである。



---

例えばロジスティック回帰だと、scikit-learn上パラメータ$C$でペナルティを調整することができる。  
ここで、$C$は$\lambda$の逆数として与えられるので、$C$が大きくなるほどペナルティが少なくなりモデルの複雑さが上がる。  
また$C$が小さくなるほどペナルティが大きくなり、モデルの複雑さはおさえられる。

```python
sample_lr = LogisticRegression(penalty='l2', C=100.0, random_state=1)
sample_lr
```

### 検証曲線によるバイアスとバリアンスの診断

ロジスティック回帰の逆正則化パラメータ$C$を変化させ、そのときのトレーニングとテストの正解率をグラフとしてプロットする。このようにモデルのパラメータの関数としてトレーニングとテストの正解率をプロットしたものを**検証曲線**という。

```python
# 検証曲線のプロット
param_range = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]

# max_iterを1000以上に設定すれば、FutureWarningがでない
pipe_lr = make_pipeline(StandardScaler(),
                        LogisticRegression(penalty='l2', random_state=1,solver='lbfgs',max_iter=1000))

train_scores, test_scores = validation_curve(
                estimator=pipe_lr, 
                X=X_train, 
                y=y_train, 
                param_name='logisticregression__C', 
                param_range=param_range,
                cv=10)

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.plot(param_range, train_mean, 
         color='blue', marker='o', 
         markersize=5, label='training accuracy')

plt.fill_between(param_range, train_mean + train_std,
                 train_mean - train_std, alpha=0.15,
                 color='blue')

plt.plot(param_range, test_mean, 
         color='green', linestyle='--', 
         marker='s', markersize=5, 
         label='validation accuracy')

plt.fill_between(param_range, 
                 test_mean + test_std,
                 test_mean - test_std, 
                 alpha=0.15, color='green')

plt.grid()
plt.xscale('log')
plt.legend(loc='lower right')
plt.xlabel('Parameter C')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.0])
plt.tight_layout()
plt.show()
```
