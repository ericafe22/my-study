### データセットの用意

```python
# 必要そうなパッケージをimport
from sklearn import datasets
import numpy as np
import pandas as pd
from scipy import interp
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import learning_curve, validation_curve
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc
from sklearn.pipeline import make_pipeline
from pydotplus import graph_from_dot_data
from IPython.display import Image
```

```python
# サンプルのIrisデータセットをロード
iris = datasets.load_iris()
# データセットの特徴を確認
print(iris.DESCR)
```

```python
# 3, 4列目の特徴量を抽出
X = iris.data[:, [2, 3]]
# クラスラベルを取得
y = iris.target
# 一意なクラスラベルを出力
# 0->Iris-setosa, 1->Iris-versicolor, 2->Iris-virginica
print('Class labels:', np.unique(y))
```

### 学習データとテストデータの分割

```python
# トレーニングデータとテストデータに分割
# 全体の30%をテストデータとする
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1, stratify=y)

print(len(X_train))
print(len(X_test))
```

# ロジスティック回帰とは？
分類(classfication)のための線形モデル  
あるクラスに属する確率を出力する  
線形分離可能な分類にのみ高い性能を発揮 

オッズ比の対数のロジット関数を定義。その逆関数がロジスティックシグモイドにあたる。  
このシグモイド関数を活性化関数として用い、重みを更新するするモデルをロジスティック回帰という。




例えばあるサンプルに対してロジスティック回帰を行い ϕ(z)=0.8 であったならば、そのサンプルがあるクラスに属するは80%である。  
このように確率を計算し閾値によって分類を行うことで、分類を行うことができる。 例えば下記のようにできる。
  
ϕ(z)≥0.5の場合は1   
ϕ(z)<0.5の場合は0

```python
# シグモイド関数の図示
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)

plt.plot(z, phi_z)
plt.axvline(0.0, color='k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')

# y axis ticks and gridline
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)

plt.tight_layout()
plt.show()
```

### sklearnを用いたロジスティック回帰モデルの実装

```python
# モデルの学習
# Cは正則化のためのパラメーター。大きくすると正則化が弱くなる
# 正則化：モデルが学習する際に複雑さが増すことに対するペナルティを設け、過学習を防ぐ(正則化し過ぎても性能がでない)
# solver：最適解の探索手法を newton-cg，lbfgs，liblinear，sag から選択する．指定しないとFutureWarningが表示される
# multi_class：多クラスを分類する場合において，あるインスタンスを対象のクラスにいれるか，それ以外に入れるか (one-versus-rest, ovr) 
#              同じく指定しないとFutureWariningが表示される
lr = LogisticRegression(C=100.0, random_state=1,solver='lbfgs', multi_class='ovr')
lr.fit(X_train_std, y_train)
```

```python
# スコア(決定係数)
print('Train score: {:.4f}'.format(lr.score(X_train_std, y_train)))
print('Test score: {:.4f}'.format(lr.score(X_test_std, y_test)))
```

```python
# 混同行列
# 45個のデータが真の値に対してどう分類されたか？
# 縦軸が正解、横軸が予測値
print('Confusion matrix:\n{}'.format(confusion_matrix(y_test, lr.predict(X_test_std))))
```

```python
# シグモイド関数をかました結果の予測値
# 出力値を分かりやすくみるため一部のデータ(3行のみ)を予測を行う
# 確率が一番大きいクラスに分類されていることが分かる
print('テストデータ')
print(X_test_std[:3, :])
print('予測される確率')
print(lr.predict_proba(X_test_std[:3, :]))
print('予測されるクラス')
print(lr.predict(X_test_std[:3, :]))
```

```python
# 結果を図示
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        # plot all samples
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')


X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))

plot_decision_regions(X=X_combined_std, y=y_combined,
                      classifier=lr, test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# 先ほどのIrisの分類でどれだけテストデータと予測がずれているか確認
y_pred = lr.predict(X_test_std)
print(len(y_test))
print(y_test == y_pred)
```

```python
# 正解率
accuracy_score(y_test, y_pred)
```
